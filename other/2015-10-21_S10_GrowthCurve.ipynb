```python
%pylab inline
import pandas as pd
```

```python
data = pd.read_csv('./data/S10_GrowthCurve.csv')
data = data.set_index(['initial cell density (10^-4 OD)', 'ampicillin (ug/ml)'])
data = data.to_panel()
```

```python
data.iloc[:, 5].T.plot()
legend(loc='center left', bbox_to_anchor=(1.0, 0.5), bbox_transform=gcf().transFigure, frameon=False,
      title='ampicillin (ug/ml)')
ylabel('Optical Density (600 nm)')
ylim(-0.05, 1.6)
grid(True)
xlabel('Time (hr)')
```
